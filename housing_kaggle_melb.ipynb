```python
from sklearn.model_selection import train_test_split
```

```python
# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
melb_model = DecisionTreeRegressor()
melb_model.fit(train_X, train_y)

predicts = melb_model.predict(val_X)
print(mean_absolute_error(val_y, predicts))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    
    pred_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, pred_val)
    return mae
```

```python
missing_cols = [col for col in train_X.columns if train_X[col].isnull().any()]

def get_mae_score(train_X, val_X, train_y, val_y):
    model = RandomForestRegressor(random_state=0)
    model.fit(train_X, train_y)
    
    pred_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, pred_val)
    return mae
```

```python
reduced_train_X = train_X.drop(missing_cols, axis = 1)
reduced_val_X = val_X.drop(missing_cols, axis=1)
print(get_mae_score(reduced_train_X, reduced_val_X, train_y, val_y))
```

```python
from sklearn.impute import SimpleImputer

simple_imputer = SimpleImputer()
imp_train_X = pd.DataFrame(simple_imputer.fit_transform(train_X))
imp_val_X = pd.DataFrame(simple_imputer.transform(val_X))

imp_train_X.columns = train_X.columns
imp_val_X.columns = val_X.columns
print(get_mae_score(imp_train_X, imp_val_X, train_y, val_y))
```

```python
dt = (train_X.dtypes == 'object')
obj_cols = list(dt[dt].index)

print(obj_cols)
```

```python
def get_mae_score(train_X, val_X, train_y, val_y):
    model = RandomForestRegressor(random_state=0)
    model.fit(train_X, train_y)
    
    pred_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, pred_val)
    return mae
```

```python
drop_train_X = train_X.select_dtypes(exclude='object')
drop_test_X = val_X.select_dtypes(exclude='object')

print(get_mae_score(drop_train_X, drop_test_X, train_y, val_y))
```

```python
from sklearn.preprocessing import LabelEncoder

label_train_X = train_X.copy()
label_val_X = val_X.copy()

label_enc = LabelEncoder()

for col in obj_cols:
    label_train_X[col] = label_enc.fit_transform(train_X[col])
    label_val_X[col] = label_enc.transform(val_X[col])
    
print(get_mae_score(label_train_X, label_val_X, train_y, val_y))
```

```python
from sklearn.preprocessing import OneHotEncoder

OH_enc = OneHotEncoder(sparse=False, handle_unknown='ignore')
oh_cols_train = pd.DataFrame(OH_enc.fit_transform(train_X[obj_cols]))
oh_cols_test = pd.DataFrame(OH_enc.transform(val_X[obj_cols]))

oh_cols_train.index = train_X.index
oh_cols_test.index = val_X.index

num_X_train = train_X.drop(obj_cols, axis=1)
num_X_valid = val_X.drop(obj_cols, axis=1)

oh_X_train = pd.concat([num_X_train, oh_cols_train], axis=1)
oh_X_val = pd.concat([num_X_valid, oh_cols_test], axis=1)

print(get_mae_score(oh_X_train, oh_X_val, train_y, val_y))
```
